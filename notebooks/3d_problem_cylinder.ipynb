# Three dimensional skyrmion structure

Three dimensional standard problem for the *Proposal for a micromagnetic standard problem for materials with Dzyaloshinskii-Moriya interaction* paper.

System:


    * An FeGe cylinder of 180 nm diameter and 20 nm thickness (along z direction)
    * Interactions: Exchange:                    8.78e-12  J m**-1
                    DMI:                         1.58e-3   J m**-2
                    Saturation Magnetisation:    3.84e5    A m**-1
                    

Here we observe a helical rotation along the thickness of the sample, which modifies the skyrmionic structure in every layer.


Relevant paper:

[x] Three-dimensional skyrmion states in thin films of cubic helimagnets. PRB **87**, 094424 (2013).

# Material

```python
# System parameters for the finite difference simulations
# Since we use a 3x3x2 nm^3 cell, we use a slightly larger sample size to get
# the right sample dimensions

# Scale in nm
side_length_s = 183
thickness_s = 21
cell_s = 1
cellz_s = 1

# scales in m
nm = 1e-9
side_length = side_length_s * nm
thickness = thickness_s * nm
cell = cell_s * nm
cellz = cellz_s * nm

# Bulk
A = 8.78e-12
D = 1.58e-3
Ms = 3.84e5
B = 0.4

initial_sk_diam = 20
```

Relevant parameters: Helix period $L_{D}$ and saturation field $H_{D}$

$$L_{D} = \frac{4\pi A}{|D|} \qquad H_{D}=\frac{D^{2}}{2 A M_{\text{s}}}$$

```python
L_D = 4 * np.pi * A / D
H_D = (D ** 2) / (2 * A * Ms)

print('L_D =', L_D * 1e9, 'nm')
print('mu0 H_D =', H_D, 'T')

print('thickness / L_D = ', thickness / L_D)
print('H / H_D = ', B / H_D)
```

# Fidimag Simulation

```python
import fidimag
import fidimag.common.constant as C
```

Here we specify the Fidimag simulation with a cylinder geometry and a skyrmion as initial state:

```python
dx, dy, dz = cell_s, cell_s, cellz_s
nx, ny, nz = int(side_length_s / dx), int(side_length_s / dy), int(thickness_s / dz)
mesh = fidimag.common.CuboidMesh(dx=dx, dy=dy, dz=dz, nx=nx, ny=ny, nz=nz,
                                 unit_length=1e-9, 
                                 # periodicity=(True, True, False)
                                 )
xs, ys, zs = mesh.coordinates[:, 0], mesh.coordinates[:, 1], mesh.coordinates[:, 2]
centre_x = xs.min() + (xs.max() - xs.min()) * 0.5
centre_y = ys.min() + (ys.max() - ys.min()) * 0.5
centre_z = zs.min() + (zs.max() - zs.min()) * 0.5
# print(centre_x, centre_y)

sim = fidimag.micro.Sim(mesh, name='fidimag_3D_cyl')

# To define the nanodisk:
def Ms_fun(pos):
    x, y, z = pos[0] - centre_x, pos[1] - centre_y, pos[2] - centre_z
    if (x ** 2 + y ** 2) ** 0.5 < (183 * 0.5) and np.abs(z) < 11:
        return Ms
    else:
        return 0

sim.set_Ms(Ms_fun)

# A linear model for the skyrmion profile
def m_init(pos):
    x, y = pos[0] - centre_x, pos[1] - centre_y
    if (x ** 2 + y ** 2) ** 0.5 < initial_sk_diam:
        r = (x ** 2 + y ** 2) ** 0.5
        phi = np.arctan2(y, x) + 0.5 * np.pi
        k = np.pi / initial_sk_diam
        return (np.sin(k * r) * np.cos(phi),
                 np.sin(k * r) * np.sin(phi),
                 -np.cos(k * r))
    else:
        return (0, 0, 1)
    
sim.set_m(m_init)

sim.add(fidimag.micro.UniformExchange(A=A))
sim.add(fidimag.micro.DMI(D=D, dmi_type='bulk'))
sim.add(fidimag.micro.Zeeman(H0=(0, 0, B / C.mu_0)))
# sim.add(fidimag.micro.UniaxialAnisotropy(Ku, axis=(0, 0, 1)))
```

We relax the system to stabilise the skyrmion solution:

```python
%%capture
sim.driver.do_precession = False
sim.driver.alpha = 0.5
# sim.driver.set_tols(1e-10, 1e-10)
sim.relax(stopping_dmdt=.01)
```

We can now extract the data from spins across the center of the system for every $xy$ plane-cut along the $z$ direction 

```python
# Define variables to store all the coordinates
mx_F, my_F, mz_F = (sim.spin.reshape(-1, 3)[:, 0], 
                    sim.spin.reshape(-1, 3)[:, 1], 
                    sim.spin.reshape(-1, 3)[:, 2])

# Redefine them with respect to the centre of the system
r_fidimag = sim.mesh.coordinates
x_F, y_F, z_F = (r_fidimag[:, 0] - centre_x, 
                 r_fidimag[:, 1] - centre_y, 
                 r_fidimag[:, 2] - centre_z)

# We will mask the array to extract only the spins at y=y_mid (middle of the sample)
# and at a specific Z coordinate (i.e. a  plane cut)
xs_F, ys_F, zs_F = np.unique(x_F), np.unique(y_F), np.unique(z_F)

_filter_y_F = lambda z_i: np.logical_and(y_F == ys_F[int(len(ys_F) * 0.5)],
                                          z_F == zs_F[z_i])

# Define functions to show the data at the middle and for a given z index
# i.e. 0 --> z[0] , 1 --> z[1]
mz_fidimag = lambda z_i: mz_F[_filter_y_F(z_i)]
my_fidimag = lambda z_i: my_F[_filter_y_F(z_i)]
mx_fidimag = lambda z_i: mx_F[_filter_y_F(z_i)]
x_fidimag = lambda z_i: x_F[_filter_y_F(z_i)]
y_fidimag = lambda z_i: y_F[_filter_y_F(z_i)]
# z_fidimag = lambda z_i: z[_filter_y_F(z_i)]

# Extract cylindrical componenets of the magnetisation
phi_F = np.arctan2(y_F, x_F)
mphi_F = (-mx_F * np.sin(phi_F) + my_F * np.cos(phi_F))
mr_F = (mx_F * np.cos(phi_F) + my_F * np.sin(phi_F))

mphi_fidimag = lambda z_i: mphi_F[_filter_y_F(z_i)]
mr_fidimag = lambda z_i: mr_F[_filter_y_F(z_i)]
```

```python
# Filter across a line is with respect to this y coordinate:
print('y =', ys_F[int(len(ys_F) * 0.5)])
```

We can find the point $x$ such that $m_{z}(x)=0$, which we define as the skyrmion radius, for every slice. We show that it is consistent across the sample thickness:

```python
for i, z in enumerate(zs_F):
    f = scipy.interpolate.interp1d(x_fidimag(i), mz_fidimag(i), kind='cubic')
    # Skyrmion radius: m_z = 0
    rsk = scipy.optimize.brentq(f, 0, 50)
    
    if i == 1:
        r_sk_F = rsk

    print('At z = {:<5}, r_sk = {:.3f}'.format(z, rsk))
```

We will use the radius from the bottom layer for the calculations. The system profile at the bottom surface looks like:

```python
fm = np.logical_and(sim.Ms > 1e-3, z_F == zs_F[0])
plt.figure(figsize=(6, 6))
plt.scatter(x_F[fm], y_F[fm], c=mz_F[fm], s=5, marker='s', cmap='RdYlBu')
circle = plt.Circle((0, 0), int(r_sk_F) + 1, fill='', lw=2, ls='--')
plt.gca().add_artist(circle)
```

We plot here an overview of the system. We show the cylinrical coordinates and snapshots of the magnetisation at two different $z$ layers, and snapshots of the system with spins drawn at $r=r_{\text{sk}}$, which is defined as the points in a layer where $m_{z}=0$:

```python
# Adjust aspect ratios according to gridspec
f = plt.figure(figsize=(8, 16 / 1.6 + 8 / 1.2))
axs = [None] * 5

# Make outer gridspec
# We will design a: 2 X 1 grid of plots, and add a 1 X 3 grid at the bottom
# The bottom plots have half the height of a  normal plot, so we set
# 2 + 2 + 1 in heights  -> 5:1 ratio
outer = gridspec.GridSpec(2, 1, height_ratios = [5, 1])
# Separation of the bottom row of plots from the top grid
outer.update(hspace=0.15)

# Make nested gridspecs
# gs2 will be used to plot snapshots of three cut planes
# We set 2 for the height of the 2 X 1 grid since each of these plots span
# 1 unit of plots. We set the width as 9 since we divide gs2 in 3, thus
# gs1 is divided in 9 (3 rows)
gs1 = gridspec.GridSpecFromSubplotSpec(2, 9, subplot_spec=outer[0], hspace=0.)
gs2 = gridspec.GridSpecFromSubplotSpec(1, 9, subplot_spec=outer[1], hspace=0., wspace=0)

# 2 X 1 grid of plots: heights span 1 unit and widths span 9 units
axs[1] = plt.subplot(gs1[1:, :9])
axs[0] = plt.subplot(gs1[:1, :9], sharex=axs[1])

# 1 X 3 bottom row of plots: heights span 1 unit and widths 3 units
for j, i in enumerate(range(2, 5)):
    axs[i] = plt.subplot(gs2[:, j * 3:(j + 1) * 3])
    axs[i].set_yticks([])
    axs[i].set_xticks([])

# Define markers
marks = ['o', '^', 's']
    
# -----------------------------------------------------------------------------
# Plot at z = 0 nm, which is the index 5

# m components
# Interval for interpolation:
xi = np.linspace(0, 91.5, 100)

# # A cubic spline interpolation doesn't work well if using the whole domain, 
# because of the singularity at $x=0$, thus we use a spline starting from $x=0$
x_half_index = len(x_fidimag(10)) // 2

l, = axs[0].plot(x_fidimag(5), mz_fidimag(10), label=r'$m_{z}$', marker=marks[0], ms=5, linewidth=0)
# Interpolation: A spline of order 3
Cs = scipy.interpolate.CubicSpline(x_fidimag(10)[x_half_index:], 
                                   mz_fidimag(10)[x_half_index:])
axs[0].plot(xi, Cs(xi), '-', color=l.get_color())

l, = axs[0].plot(x_fidimag(10), mr_fidimag(10), label=r'$m_{r}$', marker=marks[1], ms=5, linewidth=0)
Cs = scipy.interpolate.CubicSpline(x_fidimag(10)[x_half_index:], 
                                   mr_fidimag(10)[x_half_index:])
axs[0].plot(xi, Cs(xi), '-', color=l.get_color())

l, = axs[0].plot(x_fidimag(10), mphi_fidimag(10), label=r'$m_{\phi}$', marker=marks[2], ms=5, linewidth=0)
Cs = scipy.interpolate.CubicSpline(x_fidimag(10)[x_half_index:], 
                                   mphi_fidimag(10)[x_half_index:])
axs[0].plot(xi, Cs(xi), '-', color=l.get_color())

# Vertical line at the skyrmion radius
axs[0].axvline(x=r_sk_F, color='k', lw=1, ls='--')

axs[0].grid()
axs[0].legend(loc=0)
axs[0].set_xlim(0, 91.5)
axs[0].set_ylim(-1.1, 1.1)
axs[0].set_xticks([15 * i for i in range(7)])

# axs[0].set_xlabel(r'$x$ (nm)')
axs[0].set_ylabel(r'$m_{i}$')

axs[0].text(0.95, 0.05, r'$z=$' + '{}'.format(zs_F[10]),
            horizontalalignment='right', verticalalignment='bottom', 
            transform=axs[0].transAxes)

plt.setp(axs[0].get_xticklabels(), visible=False)

# -----------------------------------------------------------------------------
# Plot at z = -10 nm, which is the index 0

# Interval for interpolation:
xi = np.linspace(0, 91.5, 100)
# # A cubic spline interpolation doesn't work well if using the whole domain, 
# because of the singularity at $x=0$, thus we use a spline starting from $x=0$
x_half_index = len(x_fidimag(0)) // 2

l, = axs[1].plot(x_fidimag(0), mz_fidimag(0), label=r'$m_{z}$', marker=marks[0], ms=5, linewidth=0)
# Interpolation: A spline of order 3
Cs = scipy.interpolate.CubicSpline(x_fidimag(0)[x_half_index:], 
                                   mz_fidimag(0)[x_half_index:])
axs[1].plot(xi, Cs(xi), '-', color=l.get_color())

l, = axs[1].plot(x_fidimag(0), mr_fidimag(0), label=r'$m_{r}$', marker=marks[1], ms=5, linewidth=0)
Cs = scipy.interpolate.CubicSpline(x_fidimag(0)[x_half_index:], 
                                   mr_fidimag(0)[x_half_index:])
axs[1].plot(xi, Cs(xi), '-', color=l.get_color())

l, = axs[1].plot(x_fidimag(0), mphi_fidimag(0), label=r'$m_{\phi}$', marker=marks[2], ms=5, linewidth=0)
Cs = scipy.interpolate.CubicSpline(x_fidimag(0)[x_half_index:], 
                                   mphi_fidimag(0)[x_half_index:])
axs[1].plot(xi, Cs(xi), '-', color=l.get_color())

axs[1].axvline(x=r_sk_F, color='k', lw=1, ls='--')

axs[1].grid()
axs[1].legend(loc=0)
axs[1].set_xlim(-0.5, 91.5)
axs[1].set_ylim(-1.1, 1.1)

axs[1].set_xlabel(r'$x$ (nm)')
axs[1].set_ylabel(r'$m_{i}$')

axs[1].text(0.95, 0.05, r'$z=$' + '{}'.format(zs_F[0]),
         horizontalalignment='right', verticalalignment='bottom', 
         transform=axs[1].transAxes)


# -----------------------------------------------------------------------------
# Snapshots with the colour map and quiver plot

for i, layer in enumerate([0, 10, -1]):
    i += 2

    # Filter points with Ms=0 and for a specific XY-plane-cut (or Z layer)
    m_filter = z_F == zs_F[layer]
    
    # Interpolating the (mx,my) data with a Bivariate spline (5th order) is NOT a good approx but FAST
    interpx = scipy.interpolate.SmoothBivariateSpline(x_F[m_filter], y_F[m_filter], mx_F[m_filter], kx=5, ky=5)
    interpy = scipy.interpolate.SmoothBivariateSpline(x_F[m_filter], y_F[m_filter], my_F[m_filter], kx=5, ky=5)
    interpr = scipy.interpolate.SmoothBivariateSpline(x_F[m_filter], y_F[m_filter], mr_F[m_filter], kx=5, ky=5)
    
    # The splines require to evaluate points with .ev argument: 
    m_rad = []
    phis = np.linspace(0, 2 * np.pi, 15)
    for phi in phis:
        x, y = (r_sk_F + 0.75) * np.cos(phi), (r_sk_F + 0.5) * np.sin(phi)
        m_rad.append([x, y, interpx.ev(x, y), interpy.ev(x, y), interpr.ev(x, y)]) 
    m_rad = np.array(m_rad)
    
    # The inter2D function is better but performs badly at the boundary. It is good for the
    # data at r= r_sk though -> this function is VERY SLOW
    # interpx = scipy.interpolate.interp2d(x_F[m_filter], y_F[m_filter], mx_F[m_filter], kind='quintic')
    # interpy = scipy.interpolate.interp2d(x_F[m_filter], y_F[m_filter], my_F[m_filter], kind='quintic')
    # interpr = scipy.interpolate.interp2d(x_F[m_filter], y_F[m_filter], mr_F[m_filter], kind='quintic')
    
    # Radial basis functions are way better but still slow. We use the default ‘multiquadric’
    # radial function
#     interpx = scipy.interpolate.Rbf(x_F[m_filter], y_F[m_filter], mx_F[m_filter])
#     interpy = scipy.interpolate.Rbf(x_F[m_filter], y_F[m_filter], my_F[m_filter])
#     interpr = scipy.interpolate.Rbf(x_F[m_filter], y_F[m_filter], mr_F[m_filter])
    
#     m_rad = []
#     phis = np.linspace(0, 2 * np.pi, 15)
#     for phi in phis:
#         x, y = (r_sk_F + 0.75) * np.cos(phi), (r_sk_F + 0.5) * np.sin(phi)
#         m_rad.append([x, y, interpx(x, y), interpy(x, y), interpr(x, y)]) 
#     m_rad = np.array(m_rad)
    
    # Colour map with the m_z component
    axs[i].imshow(mz_F[m_filter].reshape(ys_F.shape[0], -1), 
                  cmap='Greys_r', 
                  vmin=-1, vmax=1, aspect='auto',
                  extent=[xs_F.min(), xs_F.max(), ys_F.min(), ys_F.max()])

    # Circle at the sk radius. We slightly shift it to match the mesh cells
    circle = plt.Circle((0, 0), int(r_sk_F) + 1, fill='', lw=2, ls='--')
    axs[i].add_artist(circle)
    
    # Quiver plot for spins at r = r_sk
    axs[i].quiver(m_rad[:, 0],  m_rad[:, 1], 
                  m_rad[:, 2], m_rad[:, 3],
                  # Colour according to m_r component
                  m_rad[:, 4],
                  cmap='RdYlBu',
                  clim=(-1, 1),
                  # Scale the arrows to make them visible
                  scale=0.06, scale_units='xy', angles='xy',
                  width=0.02, 
                  headwidth=3, headlength=4.5,
                  edgecolor='k', linewidth=0.5,
                  zorder=3
                  )
    
    # axs[i].set_title('z = {} nm'.format(zs[layer]), fontweight='light')

    # Set limits
    axs[i].set_xlim(-22, 22)
    axs[i].set_ylim(-22, 22)
    
    axs[i].set_xticks([])
    axs[i].set_yticks([])
    
    axs[i].text(0.5, -0.05, r'$z=$' + '{}'.format(zs_F[layer]),
                horizontalalignment='center', verticalalignment='top', 
                transform=axs[i].transAxes)
    
    # Draw a line from the skyrmion center to show what data is being 
    # plotted in axs[0] and axs[1]
    xl = np.linspace(0, 25, 20)
    axs[i].plot(xl, np.zeros_like(xl), '--C3')

# -----------------------------------------------------------------------------
# Colorbar of quiver plot to the left of the first snapshot (axes index = 2)
# Get coordinates and size of axes
box = axs[2].get_position()
# Define axes for the colourbar:
axColor = plt.axes([box.x0 - 0.02, box.y0, 
                    0.015, box.height])
cb = plt.colorbar(axs[2].get_children()[0], cax=axColor, orientation="vertical")
cb.set_ticks([-1, 0, 1])
# Put the colorbar labels to the left:
cb.ax.yaxis.set_ticks_position('left')
# Specify the colorbar label:
cb.set_label(r'$m_{r}$', fontsize=24, rotation=0, labelpad=-50, y=0.6)

# -----------------------------------------------------------------------------
# Label every plot: a, b, c, ...
for j, i in enumerate(['a', 'b']):
    axs[j].text(-0.16, 1, i, fontweight='bold', fontsize=30,
                horizontalalignment='left', verticalalignment='top', 
                transform=axs[j].transAxes)

axs[2].text(-0.46, 1.1, 'c', fontweight='bold', fontsize=30,
            horizontalalignment='left', verticalalignment='top', 
            transform=axs[2].transAxes)

plt.savefig('result_3d_cylinder.pdf', bbox_inches='tight')
```

From [PRB **87**, 094424 (2013)], we define $\mathbf{m}=(\sin\Theta\cos\Psi,\sin\Theta\sin\Psi,\cos\Theta)$ and we can use the ansatz $\Psi(z,\phi)=\phi+\pi/2+\tilde{\Psi}(z)$. 

If we specify the skyrmion radius as $r_{\text{sk}}$ such that $m_{z}(r_{\text{sk}})=0$, then at $r=r_{\text{sk}}$, $\Theta=\pi/2$, and $m_{x}=\cos\Psi$. Further, if we take the points where $y=0$ then $\phi=0$ and $\Psi=\pi/2 + \tilde{\Psi}=\arccos(m_{x}(r=r_{\text{sk}}))$.


We can plot $\arccos(m_{x})$ and locate the point where $r=r_{\text{sk}}$:

```python
f = scipy.interpolate.interp1d(x_fidimag(0), np.arccos(mx_fidimag(0)), kind='cubic')

plt.plot(x_fidimag(0), np.arccos(mx_fidimag(0)), 'o', label=r'$z=$' + '{}'.format(zs[0]))

xi = np.linspace(-80, 80, 200)
plt.plot(xi, f(xi), '-C3')
plt.axhline(y=f(-r_sk_F), ls='--', c='k', lw=1)

plt.axvline(x=r_sk_F, ls='--', c='k', lw=1)
plt.axvline(x=-r_sk_F, ls='--', c='k', lw=1)

plt.yticks([np.pi * 0.5, f(-r_sk_F)])
plt.gca().set_yticklabels([r'$\pi/2$'])

plt.axhline(y=np.pi * 0.5, ls='--', c='k', lw=1)

plt.ylabel(r'$\arccos(m_{x})$')

# plt.ylim(-1, 1)
```

From [PRB **87**, 094424 (2013)], at the surfaces $\tilde{\Psi}=\pi L/L_{D}$, hence we can compare with the $\Psi$ value from the simulation at $r=r_{\text{sk}}$

```python
# At r=r_sk , Theta=pi/2 --> Psi = arccos mx
f = scipy.interpolate.interp1d(x_fidimag(0), np.arccos(mx_fidimag(0)), kind='cubic')
# xmax = scipy.optimize.fmin(f, 0)[0]

print('Psi = arccos(mx(r=r_sk)) = ', f(-r_sk_F))
```

# OOMMF Simulation

We load the OOMMF simulation data from the `../sims/3D_OOMMF` folder and scale the magnetisation field and coordinates:

```python
base_folder = '../sims/'
omf_file = glob.glob(base_folder + '3D_OOMMF/oommf_3D_bulk_cylinder-Oxs*.omf')[0]
oommf_data = pd.read_csv(omf_file, 
                         comment='#', header=None, delim_whitespace=True)
oommf_data = oommf_data.values

# Scale the coordinates and the magnetization:
oommf_data[:, :3] *= 1e9 
oommf_data[:, 3:] /= Ms
```

Extract the coordinates and components of the magnetization. We create a `_filter_y_O` to extract the spins at a specific layer in the $z$ direction and at a line across the center of this layer (where the skyrmion is located) so we can compare the skyrmion profile at every layer. We also compute the cylindrical components of the magnetization.

```python
# Extract the coordinates and shift them with respect to the mesh center
r_oommf = oommf_data[:, :3]
x, y, z = r_oommf[:, 0], r_oommf[:, 1], r_oommf[:, 2]
centre_x = x.min() + (x.max() - x.min()) * 0.5
centre_y = y.min() + (y.max() - y.min()) * 0.5
centre_z = z.min() + (z.max() - z.min()) * 0.5

x_O, y_O, z_O = (r_oommf[:, 0] - centre_x,
                 r_oommf[:, 1] - centre_y,
                 r_oommf[:, 2] - centre_z)

# Magnetization coordinates of the whole mesh
mx_O, my_O, mz_O = oommf_data[:, 3], oommf_data[:, 4], oommf_data[:, 5]

# We will mask the array to extract only the spins at y=y_mid (middle of the sample)
xs_O, ys_O, zs_O = np.unique(x_O), np.unique(y_O), np.unique(z_O)
_filter_y_O = lambda z_i: np.logical_and(y_O == ys_O[int(len(ys_O) * 0.5)], z_O == zs_O[z_i])

# Functions for the coordinates and magnetization extracted for a specific layer
mz_oommf = lambda z_i: mz_O[_filter_y_O(z_i)]
my_oommf = lambda z_i: my_O[_filter_y_O(z_i)]
mx_oommf = lambda z_i: mx_O[_filter_y_O(z_i)]
x_oommf = lambda z_i: x_O[_filter_y_O(z_i)]
y_oommf = lambda z_i: y_O[_filter_y_O(z_i)]
# z_oommf = lambda z_i: z[_filter_y_oommf(z_i)]

# Same for cylindrical coordinates
phi_O = np.arctan2(y_O, x_O)
mphi_oommf = lambda z_i: (-mx_O * np.sin(phi_O) + my_O * np.cos(phi_O))[_filter_y_O(z_i)]
mr_oommf = lambda z_i: (mx_O * np.cos(phi_O) + my_O * np.sin(phi_O))[_filter_y_O(z_i)]
```

The skyrmion radius across the sample. We then use the value from the bottom layer (index 0):

```python
for i, z in enumerate(zs_O):
    f = scipy.interpolate.interp1d(x_oommf(i), mz_oommf(i), kind='cubic')
    # Skyrmion radius: m_z = 0
    rsk = scipy.optimize.brentq(f, 0, 50)
    
    if i == 0:
        r_sk_O = rsk

    print('At z = {:<6.2f}, r_sk = {:.3f}'.format(z, rsk))
```

# Theory

We compare here the solution of a 2D skyrmion in a bulk DMI disk system, with the result of the simulations for a thick cylinder. Specifically, we compare the skyrmion profile at $z=0$ because from the simulations it seems that at this $z$ there is no radial component of the magnetisation. Thus the solution resembles that of a 2D system. We will see, however, that the skyrmion profile is still affected by the thickness, even without having dipolar interactions.

By describing the magnetisation as

$$
\mathbf{m} = (\sin\theta\cos\psi,\sin\theta\sin\psi,\cos\theta) \,,
$$

where $\theta,\psi$ are n cylindrical coordinates, i.e. $\theta = \theta(r,\phi,z)$ (same for $\psi$). A minimisation equation to obtain skyrmion-like solutions in a 2D nanodisk, or for a skyrmion symmetric along the sample thickness, in a bulk system ($T$ symmetry class material), is found for $\theta=\theta(r)$ and $\Psi=\phi+\pi/2$, where $\theta$ is the solution of

$$
\begin{eqnarray}
\frac{\text{d}^2 \theta}{\text{d}r^2} & = & -\frac{1}{r} \frac{\text{d}\theta}{\text{d}r} 
                                            + \frac{1}{r^2}\frac{\sin2\theta}{2}
                                            - \frac{2\sin^{2}\theta}{\xi r} 
                                            + \beta \sin\Theta \\
\frac{\text{d}\theta}{\text{d}r} \bigg|_{r=R} & = & -\frac{1}{\xi} \,.
\end{eqnarray}
$$

Here $\beta=\mu_{0}H M_{s}/(2A)$ is the constant related to the applied field $H$ and $\xi=2A/D$ is a spatial scaling constant. The system has been defined in the $r\in[0, R]$ range. We set $\theta(r=0)=0$ (skyrmion pointing in the $+z$ direction) as initial value at the centre of the disk, hence we use a negative magnetic field to find the solution for a metastable skyrmion. We can solve this equation using the shooting method to match the condition of the derivative at the $r=R$ boundary, i.e. we try a set of $\theta'(r=0)$ initial values and find the one that satisfies the condition at $r=R$.

```python
# Use units in the nm scale

# Negative field for the ODE solution
B_n = -B

# delta = (np.sqrt(A / Ku)) / nm
xi = (2 * A / D) / nm  # spatial scale
beta = (B_n * Ms / (2 * A)) * (nm ** 2)  # reduced field constant

r_array = np.linspace(1e-15, 183 * 0.5, 500)

def ode_2d(theta, r):
    '''
    ODE for boundary value problem 
    (function given at r = 0, derivative of function given at r = R )
    '''
    return np.array([theta[1],
                     -theta[1] / r + 0.5 * np.sin(2 * theta[0]) * ((1 / r ** 2)) 
                     - 2 * (np.sin(theta[0]) ** 2) / (xi * r)]
                     + beta * np.sin(theta[0])
                     )


# Define a function to solve the differential equation in the (0, R) range
# for any theta'(r=0) condition. We need theta'(r=R) = -1/xi
# The function returns the difference of theta' with respect to the desired solution at r=R
# The shooting method only seems to work with theta(r=0)=0 as initial
# condition, and a negative field
def solve_ode(a, z):
    solution, info = scipy.integrate.odeint(ode_2d, [0, a], r_array, full_output=True)
    return solution[:, 1][-1] - (-1 / xi)
```

```python
# NBVAL_IGNORE_OUTPUT
plt.figure(figsize=(8, 8 / 1.6))

aa = np.linspace(-.5, .5, 1000)
guesses = [solve_ode(a, 0) for a in aa]

plt.plot(aa, guesses)
plt.axhline(y=0, ls='--', lw=1, color='k')
plt.ylabel(r"$\delta\theta'(r=R)$")
plt.xlabel(r"$\theta'(0)$")
plt.show()
```

```python
guess = []
guess.append(scipy.optimize.brentq(lambda a: solve_ode(a, 0), -0.3, .0))
print('Guesses = ', guess)

solutions = []
for g in guess:
    solution, info = scipy.integrate.odeint(ode_2d, [0, g], r_array, full_output=True)
    solutions.append(solution)
```

The solution for the $\Theta$ angle and $m_{z}=\cos\Theta$ look like:

```python
styles = ['-', '--', '-.', ':']
f = plt.figure(figsize=(16, 8/1.6))

ax = f.add_subplot(121)
for i, s in enumerate(solutions):
    ax.plot(r_array, s[:, 0], ls=styles[i], lw=2)
ax.set_ylabel(r"$\theta(r)$")
ax.set_xlabel(r"$r$ (nm)")

ax2 = f.add_subplot(122)
for i, s in enumerate(solutions):
    ax2.plot(r_array, np.cos(s[:, 0]), ls=styles[i], lw=2)
ax2.set_ylabel(r"$\cos\theta(r)$")
ax2.set_xlabel(r"$r$ (nm)")
```

In [PRB **87**, 094424 (2013)] an approximate solution for the 3D system is found by using a linear ansatz for the $\Psi$ angle:

$$
\Psi = \phi + \pi/2 + \tilde{\Psi}(z) \,,
$$

where $\tilde{\Psi}(z)$ sets the radial modulation of the magnetization across the thickness. Using this model, at $z=0$ (middle of the sample) the solution of $\Theta$ is equivalent to that of the 2D model, neverthless we saw that the skyrmion profile does not match the simulation results. 

Qualitatively, the linear model for $\Psi$ still describes some properties of the system. For example, there is no azimuthal modulation at the middle layer, at the bottom and top surfaces the radial modulation of $\mathbf{m}$ is maximal and it seems $\Theta$ does not depends on $z$, because $m_{z}=\cos\Theta$ is symmetric across the whole thickness. Therefore, a good solution might still be described by $\theta=\theta(r)$ and $\Psi=\Psi(\phi,z)$. It is likely that the $z$ modulation adds an extra constant in $\Theta$ so the skyrmion radius is slightly smaller than the 2D case.

We must also notice that in our case we use a field magnitude that is above the critical conical saturation field $H_{D}$, thus the linear model for the $\Psi$ angle might not be justified in this case
